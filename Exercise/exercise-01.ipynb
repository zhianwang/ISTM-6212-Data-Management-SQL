What is the difference between the two previous cells, and what does the single dot mean?

The first command 'ls -aF' shows us all the file and directory names in current directory and doesn't ignore directories starting with '.'. And 'F' append '/' to the names of directories.
The second command 'ls -af .' also shows all the file and directory names, while it shows without '/' after directories and all names are shown all in one color if you run it in Terminal. The order of the names is different for two commands as well.
The single dot '.' stands for the curent working directory.

```python
!ls -F ..
```

What do the double dots mean?

The double dots '..' mean the parent directory of the current working directory, one level up the current directory.
